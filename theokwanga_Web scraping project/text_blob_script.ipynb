```python
from textblob import TextBlob
from textblob.sentiments import NaiveBayesAnalyzer

infile = 'C:/Users/OLYMPIA/Desktop/PYTHON_JUPYTER/titles_only.csv'

with open(infile, 'r+') as csvfile:
    with open('titles_sentiment.csv', 'w') as csvnew:
        rows = csv.reader(csvfile)
        for row in rows:
            sentence = row[0]
            blob = TextBlob(sentence, analyzer=NaiveBayesAnalyzer())
            csvnew.write(sentence + " , " + str(blob.sentiment) + "\n")
            print(sentence + ", " + str(blob.sentiment)) 
# print(sentence)
```
